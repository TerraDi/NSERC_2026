```python
import numpy as np

src = "/home/dtian/scratch/Di_Research/CNN_Weights/target_maps_test.npy"         
dst = "/home/dtian/scratch/Di_Research/CNN_Weights/target_maps_test_float_16.npy"

src_arr = np.load(src, mmap_mode="r")
N = src_arr.shape[0]
print("shape", src_arr.shape, "dtype", src_arr.dtype, flush=True)

sample = src_arr[::500]                     
print("min", float(sample.min()), "max", float(sample.max()), flush=True)

dst_arr = np.lib.format.open_memmap(dst, mode="w+", dtype=np.float16, shape=src_arr.shape)
chunk = 200
for i in range(0, N, chunk):
    j = min(i + chunk, N)
    dst_arr[i:j] = src_arr[i:j].astype(np.float16)
    print(f"{j}/{N}", flush=True)
dst_arr.flush(); del dst_arr, src_arr
print("done", flush=True)
```
